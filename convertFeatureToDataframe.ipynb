```python
lists = ['CHAPLATE', 'NOTTYPE', 'MMDRIVE', 'THCARRY', 'CSCARRY', 'QSHE']
```

```python
def word_exactor(lists, dict_data):
    count = 0
    result = []
    label = np.ones(10)
    for i in lists:
        dict_test = dict_data[i]
        labels = label * count
        dict_test = np.c_[dict_test,labels]
        count += 1
        result.append(dict_test)
    result = np.array(result)
    result = result.reshape(60,19) # 7 = 3*2+1   folmuar = f*2 + 1 60 = 6 words * 10 sample
    return result
```

```python
result_2 = word_exactor(lists, dict_2)
result_4 = word_exactor(lists, dict_4)
result_5 = word_exactor(lists, dict_5)
result_6 = word_exactor(lists, dict_6)
result_7 = word_exactor(lists, dict_7)
result_8 = word_exactor(lists, dict_8)
result_2.shape
```

```python
import pandas as pd
array = np.vstack((result_2,result_4,result_5,result_6,result_7,result_8))
df = pd.DataFrame(array)
df
```
